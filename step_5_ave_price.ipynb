```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import random

df = pd.read_csv("corrected.csv",encoding="cp1252")
```

```python
mod_int = int(input("Please enter your percentage, without %"))

average_price = df["Cars Prices"].mean()
lower_bound = average_price * (1.0-(mod_int/100))
upper_bound = average_price * (1.0+(mod_int/100))

cars_in_range = df[(df['Cars Prices'] >= lower_bound) & (df['Cars Prices'] <= upper_bound)]
top_speed_cars = cars_in_range.sort_values(by='Total Speed', ascending=False).head(10)

top_speed_cars_sorted = top_speed_cars.sort_values(by='Total Speed', ascending=True)

plt.figure(figsize=(10,4))

sns.barplot(
    data=top_speed_cars_sorted,
    x="Cars Names",
    y="Total Speed",


)
plt.subplots_adjust(bottom=0.25)
plt.xticks(rotation=45)
plt.title(f"Top 10 Fastest Cars in range: {lower_bound:.2f},{upper_bound:.2f}, {mod_int} Percent range")
plt.ylabel("Top Speed")
plt.xlabel("Car Names")
plt.savefig("plots/topspeed", dpi=300, bbox_inches="tight")
plt.show()





```
