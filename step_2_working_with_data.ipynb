# Finding Averege price per company

```python
import pandas as pd
import numpy as np
import re
import seaborn as sns
import matplotlib.pyplot as plt
import random

df = pd.read_csv("corrected.csv",encoding="cp1252")
df
```

```python
# Standardize company names
pd.set_option('display.float_format', '{:,.2f}'.format) # Fix scientific notation issue
df["Company Names"] = df["Company Names"].str.upper().str.strip()

# Count of cars per company
cars_per_company = df.groupby("Company Names")["Cars Names"].count()

# Sum of car prices per company
total_price_per_company = df.groupby("Company Names")["Cars Prices"].sum()

# Average price per company = total / count
avg_price_per_company = (total_price_per_company / cars_per_company).reset_index()
avg_price_per_company.columns = ["Company Names", "Average Price"]

print(avg_price_per_company)
```

# Finding averege acceleration

```python
pet_accel = df.query("`Fuel Types` == 'Petrol'")["Performance(0 - 100 )KM/H"].mean()
bat_accel = df.query("`Fuel Types` == 'Electric'")["Performance(0 - 100 )KM/H"].mean()

print("Petrol acceleration:", round(pet_accel,2), "Electric Acceleration", round(bat_accel,2)) 

```
